```python
df = df.dropna(subset=["Products.name"]).reset_index(drop=True)
```

```python
grouped=df.groupby('Products.name').agg({'Orders.total':[np.size,np.sum,np.mean]})
```

```python
grouped.head()
```

```python
popular=grouped.sort_values(("Orders.total","sum"),ascending=False)
```
