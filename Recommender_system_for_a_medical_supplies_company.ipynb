### Matris factorization

```python
#libraries
```

```python
from surprise import Dataset, Reader
from surprise.model_selection import train_test_split
from surprise import SVD
```

```python
#preapare reader and data
```

```python
reader = Reader(rating_scale=(df['Orders.total'].min(), df['Orders.total'].max()))
data = Dataset.load_from_df(df[['Customers.id', 'Products.id', 'Orders.total']], reader)
```

```python
# Split the data into train and test sets
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)
```

```python
# Create and train the matrix factorization model (SVD)
model = SVD()
model.fit(trainset)
```

```python
# Make predictions for all pairs (customer, product) that are NOT in the training set
testset = trainset.build_anti_testset()
predictions = model.test(testset)
```

```python
# Get top N recommendations for each user
from collections import defaultdict
```

```python
def get_top_n(predictions, n=10):
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))

    # Sort the predictions for each user and retrieve the top N
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]

    return top_n
```

```python
# Get top N recommendations
top_n = get_top_n(predictions, n=5)
```

```python
# Display top N recommendations for each user
# Rewrite the code without using for loop
top_n_formatted = [
    (f"Customer ID: {uid}\n" + "\n".join([f"Product ID: {iid}, Estimated Rating: {est}" for iid, est in user_ratings]) + "\n")
    for uid, user_ratings in top_n.items()
]

# Print the results
print("\n".join(top_n_formatted))
```

### Cosine similarity for product description

```python

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
data=df
# Extract relevant columns
product_data = data[['Products.id', 'Products.short_description']].dropna()

# Create a CountVectorizer to convert text data into a bag-of-words representation
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(product_data['Products.short_description'])

# Calculate cosine similarity
cosine_sim = cosine_similarity(X, X)

# Display the cosine similarity matrix
cosine_sim_df = pd.DataFrame(cosine_sim, index=product_data['Products.id'], columns=product_data['Products.id'])

from IPython.display import display, HTML

# Set max display rows and columns
display(HTML("<style>div.output_scroll { height: 44em; }</style>"))
pd.set_option("display.max_rows", 500)
pd.set_option("display.max_columns", 500)
pd.set_option("display.width", 1000)

#print(cosine_sim_df)
```

### Company with a maximum purchase

```python
# Group data by 'Customers.company' and calculate the sum of 'Orders.total' for each company
company_purchase = data.groupby('Customers.company')['Orders.total'].sum()

# Find the company with the maximum purchase
max_purchase_company = company_purchase.idxmax()

# Display the company with the maximum purchase
print(f"Company with Maximum Purchase: {max_purchase_company}")
```
